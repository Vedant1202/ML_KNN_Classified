```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('Classified Data', index_col=0)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sns.heatmap(df.isnull())
```

```python
sns.pairplot(df)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features
```

```python
df_feature = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
df_feature
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x = df_feature
y = df['TARGET CLASS']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train, y_train)
```

```python
predictions = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
error_rate = []

for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    predictions_i = knn.predict(x_test)
    error_rate.append(np.mean(predictions_i != y_test))
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(1, 50), error_rate, color='black', animated=True, linestyle='--', marker='o', markerfacecolor='red', markersize=10)
plt.title('Elbow Method for finding optimum K')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn_new = KNeighborsClassifier(n_neighbors=40)
```

```python
knn_new.fit(x_train, y_train)
```

```python
predictions_new = knn_new.predict(x_test)
```

```python
print(confusion_matrix(y_test, predictions_new))
```

```python
print(classification_report(y_test, predictions_new))
```
